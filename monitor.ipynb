```python
import os
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
in_dir = 'checkpoints/Outdoor_nyu_wsupervised/web/images'
out_dir = '/home/ec2-user/github/Synthetic2Realistic/output'
```

```python
os.listdir(in_dir)
```

```python
def get_result(epo): 

    fig = plt.figure(figsize=(15,10))
    fig.suptitle(f'Result for {epo}')

    plt.subplot(241)
    arr = cv2.imread(os.path.join(in_dir, f'{epo}_img_t.png'))[:,:,::-1]
    plt.imshow(arr)
    plt.title('real image (input)')

    plt.subplot(242)
    arr = cv2.imread(os.path.join(in_dir, f'{epo}_img_t2t.png'))[:,:,::-1]
    plt.imshow(arr)
    plt.title('real image (reconstruction)')

    plt.subplot(243)
    arr = cv2.imread(os.path.join(in_dir, f'{epo}_lab_t_g.png'))[:,:,::-1]
    plt.imshow(arr)
    plt.title(f'depth prediction {arr.shape}')

    # synthetic
    plt.subplot(245)
    arr = cv2.imread(os.path.join(in_dir, f'{epo}_img_s.png'))[:,:,::-1]
    plt.imshow(arr)
    plt.title('synthetic image (input)')

    plt.subplot(246)
    arr = cv2.imread(os.path.join(in_dir, f'{epo}_lab_s.png'))[:,:,::-1]
    plt.imshow(arr)
    plt.title(f'synthetic label {arr.shape}')

    plt.subplot(247)
    arr = cv2.imread(os.path.join(in_dir, f'{epo}_img_s2t.png'))[:,:,::-1]
    plt.imshow(arr)
    plt.title('synthetic2real image')

    plt.subplot(248)
    arr = cv2.imread(os.path.join(in_dir, f'{epo}_lab_s_g.png'))[:,:,::-1]
    plt.imshow(arr)
    #print(arr.shape)
    plt.title(f'depth prediction {arr.shape}')

    plt.savefig(os.path.join(out_dir, f'{epo}.png'))
    plt.show()
```

```python
get_result('epoch009')
```
